```python
import pandas as pd
import numpy as np
from scipy.stats import truncnorm

# 1. 전처리된 데이터 불러오기
df = pd.read_csv('preprocessed_consumption.csv')

# 2. 생성할 데이터 개수 설정
num_samples = 1000

# 3. Amount(금액) 통계 정보
amount_mean = df['AvgAmount'].mean()
amount_std = df['AvgAmount'].std()

# 4. 음수 없는 정규분포 샘플링 (truncnorm)
a, b = (0 - amount_mean) / amount_std, np.inf
synthetic_amount = truncnorm(a=a, b=b, loc=amount_mean, scale=amount_std).rvs(num_samples).astype(int)

# 5. Description(카테고리) 실제 분포 반영 샘플링
synthetic_description = np.random.choice(
    df['Description'].unique(),
    size=num_samples,
    p=df['Description'].value_counts(normalize=True).values
)

# 6. Date, Sex 등도 실제 분포 반영 샘플링
synthetic_date = np.random.choice(df['Date'].unique(), size=num_samples)
synthetic_sex = np.random.choice(df['Sex'].unique(), size=num_samples)

# 7. Age: 20~29세 무작위 숫자 생성 (중복 허용)
ages_20s = list(range(20, 30))
synthetic_age = np.random.choice(ages_20s, size=num_samples, replace=True)

# 8. 데이터프레임 합치기
synthetic_df = pd.DataFrame({
    'Date': synthetic_date,
    'AvgAmount': synthetic_amount,
    'Description': synthetic_description,
    'Sex': synthetic_sex,
    'Age': synthetic_age  # ← 수정된 부분
})

# 9. 결과 확인 (음수 없는지 체크)
print("음수 금액 개수:", (synthetic_df['AvgAmount'] < 0).sum())
print("\n생성된 나이 분포:")
print(synthetic_df['Age'].value_counts().sort_index())

# 10. CSV로 저장
synthetic_df.to_csv('synthetic_consumption.csv', index=False)
```

```python
import numpy as np
from scipy.stats import truncnorm
from datetime import datetime, timedelta

def generate_synthetic_data(num_users=50, days=30, transactions_per_day=5):
    # 사용자 기본 정보 생성
    user_ids = np.arange(1, num_users+1)
    user_ages = np.random.choice(range(20, 30), num_users)
    user_sexes = np.random.choice(['M', 'F'], num_users)

    # 날짜 범위 설정
    start_date = datetime(2024, 4, 1)
    dates = [(start_date + timedelta(days=i)).strftime('%Y-%m-%d')
            for i in range(days)]

    # 금액 분포 설정 (원본 데이터 기반)
    amount_mean = original_df['AvgAmount'].mean()
    amount_std = original_df['AvgAmount'].std()
    a, b = (0 - amount_mean)/amount_std, np.inf

    records = []
    for uid, age, sex in zip(user_ids, user_ages, user_sexes):
        for date in dates:
            for _ in range(transactions_per_day):
                # 금액 생성 (음수 방지)
                amount = int(truncnorm(a, b, loc=amount_mean, scale=amount_std).rvs(1))

                # 카테고리 생성 (원본 분포 반영)
                category = np.random.choice(categories, p=probabilities)

                records.append({
                    'UserID': uid,
                    'Date': date,
                    'AvgAmount': amount,
                    'Category': category,
                    'Age': age,
                    'Sex': sex
                })

    return pd.DataFrame(records)

# 데이터 생성 및 저장
synthetic_df = generate_synthetic_data()
synthetic_df.to_csv('enhanced_synthetic_data.csv', index=False)

# 검증
print("\n생성 데이터 카테고리 분포:")
print(synthetic_df['Category'].value_counts(normalize=True))
```

```python
import pandas as pd
import numpy as np
from scipy.stats import truncnorm
from datetime import datetime, timedelta

# 원본 데이터 로드 (실제 경로로 변경 필요)
original_df = pd.read_csv('preprocessed_consumption.csv')

# 카테고리별 현실적 금액 범위 수동 설정 (단위: 원)
CATEGORY_LIMITS = {
    '커피/음료': {'min': 1500, 'max': 20000},
    '외식': {'min': 8000, 'max': 150000},
    '의류/의복': {'min': 10000, 'max': 200000},
    '쇼핑': {'min': 5000, 'max': 300000},
    '교통': {'min': 1250, 'max': 50000},
    '교육': {'min': 50000, 'max': 300000}  # 20대 기준
}

# 카테고리 분포 추출
category_dist = original_df['Description'].value_counts(normalize=True)
categories = category_dist.index.tolist()
category_probs = category_dist.values.tolist()

# 카테고리별 통계 계산 (수동 제약 우선 적용)
category_config = {}
for category in categories:
    subset = original_df[original_df['Description'] == category]

    # 원본 통계와 수동 제약 결합
    cfg = {
        'mean': subset['AvgAmount'].mean(),
        'std': subset['AvgAmount'].std(),
        'min': max(subset['AvgAmount'].min(), CATEGORY_LIMITS[category]['min']),
        'max': min(subset['AvgAmount'].max(), CATEGORY_LIMITS[category]['max'])
    }
    category_config[category] = cfg

# 20대 추가 제약 함수
def apply_age_constraints(row):
    amount = row['AvgAmount']
    if 20 <= row['Age'] < 30:
        if row['Category'] == '교육' and amount > 300000:
            return 300000
        if row['Category'] == '커피/음료' and amount > 15000:
            return 15000
    return amount

# 데이터 생성 함수
def generate_realistic_data(num_users=50, days=30, transactions_per_day=5):
    user_ids = np.arange(1, num_users+1)
    user_ages = np.random.choice(range(20, 30), num_users)
    user_sexes = np.random.choice(['M', 'F'], num_users)

    start_date = datetime(2024, 4, 1)
    dates = [(start_date + timedelta(days=i)).strftime('%Y-%m-%d')
            for i in range(days)]

    records = []
    for uid, age, sex in zip(user_ids, user_ages, user_sexes):
        for date in dates:
            for _ in range(transactions_per_day):
                category = np.random.choice(categories, p=category_probs)
                cfg = category_config[category]

                # 트렁케이티드 정규분포 생성
                a = (cfg['min'] - cfg['mean']) / cfg['std']
                b = (cfg['max'] - cfg['mean']) / cfg['std']
                amount = int(truncnorm(a, b, loc=cfg['mean'], scale=cfg['std']).rvs(1))

                records.append({
                    'UserID': uid,
                    'Date': date,
                    'AvgAmount': amount,
                    'Category': category,
                    'Age': age,
                    'Sex': sex
                })

    # 데이터프레임 변환 및 최종 보정
    df = pd.DataFrame(records)
    df['AvgAmount'] = df.apply(apply_age_constraints, axis=1)
    return df

# 데이터 생성 및 저장
final_df = generate_realistic_data()
final_df.to_csv('final_realistic_data.csv', index=False)

# 검증
print("카테고리별 금액 제약:")
for cat in categories:
    print(f"- {cat}: {CATEGORY_LIMITS[cat]['min']}~{CATEGORY_LIMITS[cat]['max']}원")

print("\n생성 데이터 통계:")
print(final_df.groupby('Category')['AvgAmount'].agg(['min', 'max', 'mean']))
print("\n20대 커피 최대값:", final_df[final_df['Category']=='커피/음료']['AvgAmount'].max())
print("20대 교육비 최대값:", final_df[final_df['Category']=='교육']['AvgAmount'].max())
```

```python
import pandas as pd
import numpy as np
from scipy.stats import truncnorm
from datetime import datetime, timedelta

# 1. 원본 데이터 로드 (경로를 실제 파일명으로 바꿔주세요)
original_df = pd.read_csv('preprocessed_consumption.csv')

# 2. 카테고리별 현실적 금액 범위 수동 설정 (일부만 직접 지정, 나머지는 기본값)
CATEGORY_LIMITS = {
    '커피/음료': {'min': 1500, 'max': 20000},
    '외식': {'min': 8000, 'max': 150000},
    '의류/의복': {'min': 10000, 'max': 200000},
    '쇼핑': {'min': 5000, 'max': 300000},
    '교통': {'min': 1250, 'max': 50000},
    '교육': {'min': 50000, 'max': 300000}
}
DEFAULT_MIN = 1000
DEFAULT_MAX = 100000

# 3. 원본 데이터의 모든 카테고리 추출 및 누락된 카테고리 기본값 할당
all_categories = original_df['Description'].unique().tolist()
for category in all_categories:
    if category not in CATEGORY_LIMITS:
        CATEGORY_LIMITS[category] = {
            'min': DEFAULT_MIN,
            'max': DEFAULT_MAX
        }
        print(f"⚠️ 경고: '{category}' 카테고리가 CATEGORY_LIMITS에 없어 기본값({DEFAULT_MIN}~{DEFAULT_MAX}원)으로 설정됨")

# 4. 카테고리별 통계 계산 (원본 통계와 수동 제약 결합)
category_config = {}
for category in all_categories:
    subset = original_df[original_df['Description'] == category]
    mean = subset['AvgAmount'].mean() if not subset.empty else (CATEGORY_LIMITS[category]['min'] + CATEGORY_LIMITS[category]['max']) / 2
    std = subset['AvgAmount'].std() if not subset.empty else (CATEGORY_LIMITS[category]['max'] - CATEGORY_LIMITS[category]['min']) / 4
    config = {
        'mean': mean,
        'std': std if std > 0 else (CATEGORY_LIMITS[category]['max'] - CATEGORY_LIMITS[category]['min']) / 4,
        'min': CATEGORY_LIMITS[category]['min'],
        'max': CATEGORY_LIMITS[category]['max']
    }
    category_config[category] = config

# 5. 카테고리 분포 추출
category_dist = original_df['Description'].value_counts(normalize=True)
categories = category_dist.index.tolist()
category_probs = category_dist.values.tolist()

# 6. 20대 추가 제약 함수
def apply_age_constraints(row):
    amount = row['AvgAmount']
    if 20 <= row['Age'] < 30:
        if row['Category'] == '교육' and amount > 300000:
            return 300000
        if row['Category'] == '커피/음료' and amount > 15000:
            return 15000
    return amount

# 7. 데이터 생성 함수
def generate_realistic_data(num_users=50, days=30, transactions_per_day=5):
    user_ids = np.arange(1, num_users+1)
    user_ages = np.random.choice(range(20, 30), num_users)
    user_sexes = np.random.choice(['M', 'F'], num_users)

    start_date = datetime(2024, 4, 1)
    dates = [(start_date + timedelta(days=i)).strftime('%Y-%m-%d')
            for i in range(days)]

    records = []
    for uid, age, sex in zip(user_ids, user_ages, user_sexes):
        for date in dates:
            for _ in range(transactions_per_day):
                category = np.random.choice(categories, p=category_probs)
                cfg = category_config[category]

                # 트렁케이티드 정규분포 생성
                a = (cfg['min'] - cfg['mean']) / cfg['std']
                b = (cfg['max'] - cfg['mean']) / cfg['std']
                amount = int(truncnorm(a, b, loc=cfg['mean'], scale=cfg['std']).rvs(1))

                records.append({
                    'UserID': uid,
                    'Date': date,
                    'AvgAmount': amount,
                    'Category': category,
                    'Age': age,
                    'Sex': sex
                })

    # 데이터프레임 변환 및 최종 보정
    df = pd.DataFrame(records)
    df['AvgAmount'] = df.apply(apply_age_constraints, axis=1)
    return df

# 8. 데이터 생성 및 저장
final_df = generate_realistic_data()
final_df.to_csv('final_realistic_data.csv', index=False)

# 9. 검증
print("\n카테고리별 금액 제약:")
for cat in categories:
    print(f"- {cat}: {CATEGORY_LIMITS[cat]['min']}~{CATEGORY_LIMITS[cat]['max']}원")

print("\n생성 데이터 통계:")
print(final_df.groupby('Category')['AvgAmount'].agg(['min', 'max', 'mean']))
```
